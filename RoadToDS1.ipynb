### Set Identities

Let all sets referred to below be subsets of a universal set U.
1. Commutative Laws: For all sets A and B,  
(a) A ∪ B = B ∪ A and (b) A ∩ B = B ∩ A
2. Associative Laws: For all sets A, B, and C,  
(a) (A ∪ B) ∪ C = A ∪ (B ∪ C) and (b) (A ∩ B) ∩ C = A ∩ (B ∩ C)
3. Distributive Laws: For all sets A, B,and C,  
(a) A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C) and (b) A ∩ (B ∪ C) = (A ∩ B) ∪ (A ∩ C)
4. Identity Laws: For all sets A,  
(a) A ∪ ∅ = A and (b) A ∩ U = A
5. Complement Laws: For all sets A,  
(a) A ∪ A<sup>C</sup> = U and (b) A ∩ A<sup>C</sup> = ∅
6. Double Complement Law: For all sets A,  
(A<sup>C</sup>)<sup>C</sup> = A
7. Idempotent Laws: For all sets A,  
(a) A ∪ A = A and (b) A ∩ A = A
8. Universal Bound Laws: For all sets A,  
(a) A ∪ U = U and (b) A ∩ ∅ = ∅
9. De Morgan’s Laws: For all sets A and B,  
(a) (A ∪ B)<sup>C</sup> = A<sup>C</sup> ∩ B<sup>C</sup> and (b) (A ∩ B)<sup>C</sup> = A<sup>C</sup> ∪ B<sup>C</sup>
10. Absorption Laws: For all sets A and B,  
(a) A ∪ (A ∩ B) = A and (b) A ∩ (A ∪ B) = A
11. Complements of U and ∅:  
(a) U<sup>C</sup> = ∅ and (b) ∅<sup>C</sup> = U
12. Set Difference Law: For all sets A and B,  
A \ B = A ∩ B<sup>C</sup>

## Counting

Permutations and Combinations

```python
def factorial(n):
  res = 1
  
  for i in range(1,n+1):
    res *= i
  
  return res
```

```python
print(factorial(1))
print(factorial(0))
print(factorial(2))
print(factorial(3))
print(factorial(4))
```

```python
def P(n, r):
  return int(factorial(n) / factorial(n-r))
```

```python
def C(n, r):
  return int(factorial(n) / (factorial(r) * factorial(n-r)))
```

```python
P(3,2)
```

```python
C(3,2)
```
